```python
import pydeck as pdk
```

```python
import pandas as pd
import os
import geopandas as gpd
```

```python
df = pd.read_csv("../Output/Gangseo_250m.csv")
df = df.fillna(0)
```

```python
df = df[['위도', '경도', 'econmically_lbl']]
df.head()
```

```python
# df['정규화경제인구'] = df['econmically_lbl'] / df['econmically_lbl'].max()

layer = pdk.Layer(
    'CPUGridLayer', # 대용량 데이터의 경우 'GPUGridLayer'
    df,
    get_position='[경도, 위도]',
    pickable=True,
    auto_highlight=True
)
# layer.get_fill_color = '[255*정규화경제인구, 255*정규화경제인구, 255]'
layer.cellSizePixels = 10

center = [126.986, 37.565]
view_state = pdk.ViewState(
    longitude=center[0],
    latitude=center[1],
    zoom=10)

view_state.bearing = 0
view_state.pitch = 0

r = pdk.Deck(layers=[layer], initial_view_state=view_state)
# r.show()
r.to_html('../Result/grid/demo.html')
```
